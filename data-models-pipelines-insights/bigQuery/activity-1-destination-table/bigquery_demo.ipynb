# Atividade Prática: Consultando BigQuery com Python
Este notebook demonstra como utilizar a biblioteca `google-cloud-bigquery` para executar uma consulta SQL em um conjunto de dados público do Google Cloud diretamente pelo Colab.

### 1. Autenticação
No Google Colab, precisamos nos autenticar para acessar os serviços do Google Cloud.

```python
from google.colab import auth
auth.authenticate_user()
print('Usuário Autenticado')
```

### 2. Configuração do Cliente BigQuery
Substitua `'seu-projeto-id'` pelo ID do seu projeto no Google Cloud Console.

```python
from google.cloud import bigquery

project_id = 'seu-projeto-id' # @param {type:"string"}
client = bigquery.Client(project=project_id)
```

### 3. Execução da Consulta (Extração)
Vamos extrair os 10 endereços com o maior número de árvores em São Francisco.

```python
query = """
SELECT
    address,
    COUNT(address) AS number_of_trees
FROM
    `bigquery-public-data.san_francisco_trees.street_trees`
WHERE
    address != "null"
GROUP BY address
ORDER BY number_of_trees DESC
LIMIT 10;
"""

query_job = client.query(query)
results = query_job.to_dataframe()

print("Consulta executada com sucesso!")
results
```

### 4. Análise dos Resultados
Abaixo vemos a tabela de destino gerada pela consulta SQL.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.barh(results['address'], results['number_of_trees'], color='green')
plt.xlabel('Número de Árvores')
plt.title('Top 10 Endereços com Mais Árvores em SF')
plt.gca().invert_yaxis()
plt.show()
```
